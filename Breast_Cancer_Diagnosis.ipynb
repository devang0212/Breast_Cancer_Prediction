## L1 Regularization (LASSO) to choose important features

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lasso = Lasso(alpha=0.01)
lasso.fit(X, y)
```

#### Selecting features based on non-zero coefficients and getting selected features indices

```python
model = SelectFromModel(lasso, prefit=True)
X_new = model.transform(X)

selected_features = model.get_support(indices=True)
print('Selected features indices: ', selected_features)
```

```python
df_selected = df.iloc[:, selected_features]
df_selected = pd.concat([df.iloc[:, 0], df_selected], axis=1)
df.drop(df.columns.difference(df_selected.columns), axis=1, inplace=True)
df
```

## Distribution Plot for the Dataset

## Splitting the Dataset

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Training XGBoost on training set

```python
! pip install xgboost
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

## Evaluating the model's performance

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', cbar=False)

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')

plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
